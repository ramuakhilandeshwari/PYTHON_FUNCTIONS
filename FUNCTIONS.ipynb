```python
#Method that will accept the quantity value
def quantity():
  a2=0
  while a2==0:
    try:
      qty=int(input("\nEnter the quantity:\t"))
    except ValueError:
      print("Invalid")
      continue
    if qty<=0:
      print("Invalid Choice")
      continue
    else:
      a2=1
    return qty
```

```python
#Method that will accept the item selection
def option():
  a1=0
  while a1==0:
    try:
      opt=int(input("\n\nSelect the item which you wanted to purchase(1-6):\t "))
    except ValueError:
      print("Select the avalible Product")
      continue
    if opt<=0 or opt>6:
      print("Select the available Product")
      continue
    else:
      a1=1
    return opt
```

```python
#Lambda function to print the designed line    
Dec=lambda x: x*34

#Function for Bill Payment
def payment(amt):
    print("\n\n\n")
    print(Dec("##"))
    print("\t\t\tWELCOME TO THE PAYMENT SECTION\n")
    dt=datetime.datetime.now()
    print(Dec("##"))
    print("\tDate and Time:\t", dt)
    print("\n\n")
    print(Dec("=="))
    print("|{:^30}|{:^30}|".format("AED",amt))
    print(Dec("=="))
    #The custom duties for the materials that are bought from Dubai to India will include all the takes and shipping payments also. 
    INR=amt*20.40
    print("|{:^30}|{:^30}|".format("INR",INR))
    print(Dec("=="))
    if INR>10000:
        disc=0.15*INR
    else:
        disc=0.00
    print(Dec("=="))
    print("|{:^30}|{:^30}|".format("Discount in (Rs)",disc))
    print(Dec("=="))
    CustDuties=INR*28.852
    print("|{:^30}|{:^30}|".format("Custom Duties in(Rs)",CustDuties))
    print(Dec("=="))
    Amt_Payable=INR+CustDuties-disc
    print("|{:^30}|{:^30}|".format("Amount to be Paid in (Rs)",Amt_Payable))
    print(Dec("=="))
    #After this statement, it will redirect to payment website
    
    print("\n\n\n")
    
    print("*************************|||\t\t\t ****Thank You!!!**** \t\t\t|||*************************\n")
    print("\n\n\n\n\n")
    UserInput()
```
